```python
#Basic generator code

def seq(n):
  num=0
  while True:
    yield num
    num+=n
a=int(input("enter a number:"))
gen=seq(a)
print(next(gen))
print(next(gen))
```

```python
#Comparison of list and generator

def square_list(n):
    numbers = []
    for i in range(n):
        numbers.append(i**2)
    return numbers

def square_gen(n):
    for i in range(n):
        yield i**2

n = 5

list_result = square_list(n)
print("List:", list_result)

gen_result = square_gen(n)
print("Generator (first call):", next(gen_result))
print("Generator (remaining values):", list(gen_result))
```

```python
#List stores all values in memory at once.
#Generator stores only the current value.

#List has indexes.
#Generator doesn't have indexes.

#List supports slicing and indexing.
#Generator doesn't support indexing and slicing.

#List is suitable for small and fixed datasets.
#Generator is suitable for large,streaming datasets.

#List is faster for small data
#Generator is memory efficient for big data
```
